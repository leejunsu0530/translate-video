```python
from deep_translator import GoogleTranslator

texts = [
    "え、マジで？それ今言う？",
    "いやいや、そういう意味じゃないんだけどさ。",
    "ちょっと待ってって言ったよね？",
    "もう無理。今日はここまででいいや。",
    "なんか微妙じゃない？",
    "それ、前にも言った気がする。",
    "別に怒ってるわけじゃないよ。",
    "てかさ、それ誰情報？",
    "まぁ、そういうこともあるよね。",
    "正直言うと、あんまり気が進まない。",
    "はいはい、分かりましたよ。",
    "一応聞くけど、それ本気？",
    "あとでやるって言って、結局やらないやつでしょ。",
    "そこまで言わなくてもよくない？",
    "空気読んでよ、今それ言うタイミングじゃないでしょ。"
]

translated = GoogleTranslator('ja', 'ko').translate_batch(texts)
print(translated)
```
